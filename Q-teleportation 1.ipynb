```python
from qiskit import *
```

```python
circuit=QuantumCircuit(3,3)
```

```python
%matplotlib inline
```

```python
circuit.draw(output='mpl')
```

```python
circuit.x(0)
circuit.barrier()
circuit.draw(output='mpl')
```

```python
circuit.h(1)
circuit.cx(1,2)
circuit.draw(output='mpl')
```

Now q1 and q2 are entangled

```python
circuit.cx(0,1)
circuit.h(0)
circuit.draw(output='mpl')
```

<b>int this measure means measure qubits(0,1)and put it in a classical register (0,1)</b>

```python
circuit.barrier()
circuit.measure([0,1], [0,1])
circuit.draw(output='mpl')
```

```python
circuit.barrier()
circuit.cx(1,2)
circuit.cz(0,2)
circuit.draw(output='mpl')
```

<b>this is a quantum teleportation ckt.<br>
before barrier 1 we have changed the q-state of our target bit to  1<br>
inside the barrier 2 we are now entangling the two bits q1,q2<br>
barrier 3 measures 1,0 and copies it to 2<br>
and hence teleported</b>

```python
circuit.measure(2,2)
simulator=Aer.get_backend('qasm_simulator')
result=execute(circuit,backend=simulator,shots=1024).result()
count=result.get_counts()
```

```python
from qiskit.tools.visualization import plot_histogram
plot_histogram(count)
```
