<h1>Quantum Teleportation Example 2</h1>
teleporting ( |0>+|1>)/2^0.5) 

```python
circuit1=QuantumCircuit(3,3)
circuit1.draw(output='mpl')
```

```python
circuit1.h(0)
circuit1.barrier()
circuit1.draw(output='mpl')
```

applying hadamard gate to achive 0+1/root(2) state for teleportation

```python
circuit1.h(1)
circuit1.cx(1,2)
circuit1.draw(output='mpl')
```

```python
circuit1.cx(0,1)
circuit1.h(0)
circuit1.draw(output='mpl')
```

```python
circuit1.barrier()
```

```python
circuit1.draw(output='mpl')
```

entangling bits

```python
circuit1.measure([0,1],[0,1])
circuit1.draw(output='mpl')
```

```python
circuit1.barrier()
circuit1.cx(1,2)
circuit1.cz(0,2)
circuit1.draw(output='mpl')
```
